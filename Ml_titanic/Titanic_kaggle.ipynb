```python
import re #regular expression

def extract_title(name):
  p = re.compile(r", ([\w\s]+)\.")

  return p.search(name).group(1)

train_df['Title'] = train_df['Name'].apply(lambda name: extract_title(name))
```

```python
train_df['Title'].value_counts()
```

```python
test_df['Title'] = test_df['Name'].apply(lambda name: extract_title(name))
```

```python
test_df['Title'].value_counts()
```

```python
def group_title(title):
  if title in ['Mr', 'Mrs', 'Miss', 'Master']:
    return title
  elif title == 'Ms':
    return 'Miss'
  else:
    return 'Others'

train_df['Title'] = train_df['Title'].apply(lambda title: group_title(title))
test_df['Title'] = test_df['Title'].apply(lambda title: group_title(title))
```

# **Family_sizw**
SibSp, Parch

```python
train_df['Family_Size'] = train_df['SibSp'].astype('int') + train_df['Parch'].astype('int') +1
test_df['Family_Size'] = test_df['SibSp'].astype('int') + test_df['Parch'].astype('int') +1
```

```python
train_df['Family_Cat'] = pd.cut(train_df['Family_Size'], bins = [0,1,4,6,20], labels = ['solo','small', 'medium', 'large'])
test_df['Family_Cat'] = pd.cut(test_df['Family_Size'], bins = [0,1,4,6,20], labels = ['solo','small', 'medium', 'large'])
```

```python
sns.countplot(data = train_df, x = 'Family_Cat', hue = 'Survived', palette = 'Set3');
plt.legend(title = 'Name and Survived', loc = 'upper right', labels = ('Not Survived', 'Survived'))
plt.show()
```

```python
num_features = ['Age', 'Fare']
cat_features = ['Sex', 'Embarked', 'Pclass', 'Title', 'Family_Cat']
features_cols = num_features + cat_features
print(features_cols)
```

```python
def display_missing (df, features_cols):
  n_rows = df.shape[0]
  for col in features_cols:
    missing_count = df[col].isnull().sum()
    if missing_count > 0:
      print(f'{col} has {missing_count*100/n_rows:.2f} % missing values')


display_missing(train_df, features_cols)
display_missing(test_df, features_cols)
```

# **Filling missing values **
Age
- Filling missing values with median of whole dataset

```python
age_by_sex_pclass = train_df.groupby(['Sex', 'Pclass'])['Age'].median()
```

```python
age_by_sex_pclass
```

```python
train_df['Age'] = train_df.groupby(['Sex', 'Pclass'], observed=False,group_keys=False)['Age'].apply(lambda x: x.fillna(x.median()))
test_df['Age'] = test_df.groupby(['Sex', 'Pclass'], observed=False,group_keys=False)['Age'].apply(lambda x: x.fillna(x.median()))
```

```python
display_missing(train_df, features_cols)
display_missing(test_df, features_cols)
```
